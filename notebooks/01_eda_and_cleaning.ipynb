```python
import pandas as pd
import json
import os
import sys

# Add the 'src' directory to the system path so we can import our scripts
sys.path.append(os.path.abspath(os.path.join('..', 'src')))

from preprocessing import clean_text
```

```python

# 1. Load the JSON Data
data_path = '../data/raw/Ecommerce_FAQ_Chatbot_dataset.json'

with open(data_path, 'r') as f:
    raw_data = json.load(f)

```

```python
# 2. Convert to Pandas DataFrame
# The dataset structure is usually {'questions': [...]}
# We need to inspect the keys first to be sure
print(f"Keys in JSON: {raw_data.keys()}")

# Assuming the key is 'questions' based on the standard Kaggle dataset
# Note: Adjust 'questions' below if the print statement shows a different key
df = pd.DataFrame(raw_data['questions']) 

```

```python
# 3. Apply Cleaning
# We create a new column 'clean_tokens'
df['clean_tokens'] = df['question'].apply(clean_text)
```

```python
# 4. Preview Data
print("\n--- Data Preview ---")
print(df[['question', 'clean_tokens']].head())

print(f"\nTotal Questions Loaded: {len(df)}")
```
